```python
import json
# load data using Python JSON module
with open('datauser.json','r') as f:
    data = json.loads(f.read())
# Normalizing data
multiple_level_data = pd.json_normalize(data, record_path =['data'], meta =['status'], meta_prefix='config_params_', record_prefix='dbscan_')
multiple_level_data.head(5)
```

```python
column_names = ['capacity', 'power', 'user', 'timestamp','status']
```

```python
multiple_level_data.columns = column_names
```
